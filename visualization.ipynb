```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
result = pd.read_csv("visualize.csv")
```

```python
# check that data was loaded correctly
result.tail()
```

```python
# TODO(Task 4):
# Create your own visualization based on your data and what you want to show/analyze
# Feel free to get use any plotting library you're comfortable with, just remember to add it to the requirements.txt

pd.options.plotting.backend = 'plotly'

# Evolving of training session results
result.plot(
    x="timestamp",
    y=["Success", "Fail", "Miss"],
    kind="line", 
    labels = dict(
        timestamp = "Time (d)",
        value = "Proportion (%)",
        variable = "Outcome"
    )
)
```

```python
# Comparison between results of the first and last 5 training session
fig = make_subplots(
    rows = 1, cols = 2,
    specs = [[{'type':'domain'}, {'type':'domain'}]],
    subplot_titles = ["First five", "Latest five"]
)
fig.add_trace(
    go.Pie(
        labels = ["Miss", "Fail", "Success"],
        values = result.head()[["Miss", "Fail", "Success"]].mean(),
        sort = False, name = "First five"
    ), 1, 1
)
fig.add_trace(
    go.Pie(
        labels = ["Miss", "Fail", "Success"],
        values = result.tail()[["Miss", "Fail", "Success"]].mean(),
        sort = False, name = "Last five"
    ), 1, 2
)
fig.update_layout(title_text = "Comparison of average results of training sessions")
```

## My short analysis of the result

As the first figure shows, the training appears to be effective. While the proportions of misses, fails and successes do fluctuate between sessions and there is often negative progress between subsequent sessions (i.e. number of misses/fails grows compared to last session and number of successes decreases), the peaks of the proportion of successes are consistently increasing over time, while the peaks of misses and fails keep decreasing. Overall this implies that the progress is positive over a longer period of time.

The effect of the training is even more apparent from the second figure, which shows a comparison of average results from the initial five training sessions compared to last five sessions. The percentage of successes has nearly doubled, going from 31% all the way up to 60.2%. The decrease in the percentage of misses is even more apparent: At the start of the training users were nearly three times more likely to fail when compared to the current fail rate of 3.4%. Misses still remain quite common at 36.4%, which is still a clear decrease from the initial 59.8%. More notably, success is now the most likely outcome, while initially it was missing.
